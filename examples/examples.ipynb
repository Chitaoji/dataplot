```python
n = 10000

dp.random_seed(42)

x1 = dp.randn(n)
x2 = dp.randn(n)
x3 = dp.randn(n)
x4 = dp.randn(n)
x5 = dp.randn(n)
```

```python
s = x1.join(x1 + x2, x1 + x2 + x3)
a0 = s.hist(title="Sums of i.i.d Standard Normal Distributions")
a0
```

```python
a1 = (
    x1
    .join(x2, x3, x4, x5)
    .cumsum()
    .plot(
        title="Cummulative Sums of Random Numbers",
        xlabel="count",
        ylabel="sums",
        reference_lines=[f"x={n // 2}", f"x={n * 3 // 4}"],
    )
)
a1
```

```python
a2 = x1.join(x1.exp()).qqplot()
a2
```

```python
a3 = (x1 + x2 + x3).ksplot()
a3
```

```python
a4 = x1.join(x1 + x2 + x3).ppplot()
a4
```

```python
sum4 = x1 + x2 + x3 + x4
sum5 = sum4 + x5
a5 = s.join(sum4, sum5).corrmap()
a5
```

```python
dp.figure(
    a0, a1, a2, a3, a4, a5, title="Figure Collections", subplots_adjust={"top": 0.92}
)
```
